```python
#ADAM BASHFORD N = 2
import numpy as np
import matplotlib.pyplot as plt

def AB2(f,a,b,h,x0):
  A = 3/2
  B = -1/2
  n = int((b-a)/h)
  u = np.zeros(n+1)
  t = np.array([a+i*h for i in range(n)])
  u[0]=x0
  u[1]= u[0]+A*h*f(t[0],u[0])
  for i in range(1,n-1):
    u[i+1] = u[i] + h* (A*f(t[i],u[i]) + B*f(t[i-1],u[i-1]))
  return u[n-1]

f = lambda t,y: (2-2*t*y)/(t**2+1)
a = 0
b = 1
h = 0.01
x0 = 1
g = lambda t: (2*t+1)/(t**2+1) 
#print("actual number: ", np.exp(2))
print("aprox: ", AB2(f,a,b,h,x0))
#print("error: ", np.abs(AB3(f,a,b,n,x0) - np.exp(2)))
X = np.linspace(0.01,1,99)
Y = [AB2(f,a,i,h,x0) for i in X]
Z = g(X)
plt.plot(X,Y)
plt.plot(X,Z)
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
#plt.scatter(x_data,y_data)
plt.legend(("mÃ©todo de Adams-Bashford","g(t) = (2*t+1)/(t**2+1)"))
```

```python
#ADAM BASHFORD N = 3
import numpy as np
import matplotlib.pyplot as plt

def AB2(f,a,b,h,x0):
  A = 3/2
  B = -1/2
  n = int((b-a)/h)
  u = np.zeros(n+1)
  t = np.array([a+i*h for i in range(n)])
  u[0]=x0
  u[1]= u[0]+A*h*f(t[0],u[0])
  for i in range(1,n-1):
    u[i+1] = u[i] + h* (A*f(t[i],u[i]) + B*f(t[i-1],u[i-1]))
  return u[n-1]

k = 0.1
g = 9.8
m = 70.0
a = 0.0
b = 2.5
z = 0.025
x0 = 0.0
T = lambda t,v: g - (k/m)*v**2
print("Velocidad en t = 2.5s: ", AB2(T,a,b,z,x0))
#print("Velocidad en t = 2.5s: ", AB2(T,0.0,0.075,z,x0))
X = np.linspace(0.025,b,100)
Y = [AB2(T,a,i,z,x0) for i in X]
plt.plot(X,Y)
plt.legend(("V vs t","."))
plt.xlabel("Tiempo")
plt.ylabel("Velocidad")
```
