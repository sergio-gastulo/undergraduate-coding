```python
import numpy as np 

def Ronberg(a,b,f,J):
  h = b-a
  T = np.zeros((J+1,J+1),float)
  for j in range(1,J+1):
    T[j][1] = (h/2)*(f(a)+2*np.sum(np.array([f(a+j*h) for j in range(1,2**(j-1))])) + f(b))
    for k in range(2,j+1):
      T[j][k] = T[j][k-1] + (1/(4*(k-1)-1))* (T[j][k-1]-T[j-1][k-1])
      #para corroborar: print(j, "**", k, "**",T[j][k])
    h = h/2
  return T[J][J]

f = lambda x: np.cos(x)
a = 0
b = np.pi/6
J = 6
Ronberg(a,b,f,J)
```

```python
#ERROR: -1/12 h**2 (b-a) f2
def trapezio(a,b,f,n):
  h = (b-a)/n
  t = (h/2)*(f(a)+f(b)+2*np.sum(np.array([f(a+i*h) for i in range(1,n)])))
  return t

f = lambda x: np.sin(x)
a = 0
b = np.pi/2
n = 200
trapezio(a,b,f,n)
```

```python
import numpy as np
#error -1/180 h**4 (b-a) f4
def simpson(a,b,f,n=int):
  #n must be odd
  h = (b-a)/n
  Fo = np.array([f(a+(2*i-2)*h) for i in range(2,int(n/2)+1)])
  Fe = np.array([f(a+(2*i-1)*h) for i in range(1,int(n/2)+1)])
  t = (h/3)*(f(a)+f(b)+2*np.sum(Fo) + 4*np.sum(Fe))
  return t

f = lambda x: np.sin(x)
a = 0
b = np.pi/2
n = 10
simpson(a,b,f,n)
```

```python
import numpy as np

def cheby_zeros(a,b,n):
  #Zeros of U_{n+1} (n+1 zeros)
  arr = np.array([((a+b)/2 + ((b-a)/2)*np.cos((2*i+1)*np.pi/(2*n))) for i in range(n)])
  return arr[::-1]

def vandermonde(x,n):
  return ([[a**i for i in range(n)] for a in x])

def integralofmonic(a,b,n):
  return ((b**(n+1) - a**(n+1))/(n+1))

def cuadratura_cheby(a,b,n):
  x = cheby_zeros(a,b,n)
  A = np.transpose(vandermonde(x,n))
  b = np.array([integralofmonic(a,b,i) for i in range(n)])
  return np.linalg.solve(A,b)

def integral(a,b,f,n):
  w = cuadratura_cheby(a,b,n)
  x = cheby_zeros(a,b,n)
  return np.sum(np.array([ w[i]*f(x[i]) for i in range(n)]))

f = lambda x: np.sin(x)
a = 0
b = np.pi/2
n = 9
u = integral(a,b,f,n)
print("valor aproximado:", integral(a,b,f,n))
```
